```python
# Import library yang diperlukan
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset iris dari scikit-learn
iris = load_iris()
X = iris.data

# Inisialisasi model K-Means dengan 3 kelompok (sesuai dengan jumlah kelas pada iris dataset)
kmeans = KMeans(n_clusters=3, random_state=42)

# Latih model menggunakan data tanpa label
kmeans.fit(X)

# Dapatkan label prediksi dari model
labels = kmeans.labels_

# Visualisasi hasil clustering
plt.figure(figsize=(8, 6))

plt.scatter(X[:, 0], X[:, 1], c=labels, s=100, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('Pengelompokan Data Iris dengan K-Means')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
